```python
df = df1.merge(df2, on='NOC')
```

```python
#sorting oldest top 20 events
df.sort_values('Year')[:20]
```

```python
old_sports = df[df['Year'] <= 1897]
```

```python
old_sports.groupby('Sport').Year.value_counts()
```

```python
dff = old_sports.groupby('Sport').Year.value_counts()
dff.plot(kind='bar')
plt.title('top oldest sports')
```

```python
#filtering oldest sports
sports = ['Athletics', 'Cycling', 'Fencing', 'Gymnastics', 'Shooting', 'Swimming', 'Tennis', 'Weightlifting', 'Wrestling']
df = df[(df['Sport'] == sports[0]) | (df['Sport'] == sports[1]) | (df['Sport'] == sports[2]) | (df['Sport'] == sports[3]) | (df['Sport'] == sports[4]) | (df['Sport'] == sports[5]) | (df['Sport'] == sports[6]) | (df['Sport'] == sports[7]) | (df['Sport'] == sports[8])]
```

```python
df_male = df[df['Sex'] == 'M']
df_Fmale = df[df['Sex'] == 'F']
```

```python
df_male.groupby(['NOC', 'Sport']).Medal.value_counts().sort_values(ascending=False)[:20]
```

```python
rcParams['figure.figsize']= 10, 8
sns.countplot(data=df_male, x=df_male['NOC'], hue= 'Medal');
plt.xticks(rotation='vertical');
```
