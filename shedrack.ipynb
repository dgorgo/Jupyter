```python
df = df1.merge(df2, on='NOC')
```

```python
old_sports = df[df['Year'] <= 1897]
```

```python
old_sports.groupby('Sport').Year.value_counts()
```

```python
dff = old_sports.groupby('Sport').Year.value_counts()
dff.plot(kind='bar')
plt.title('top oldest sports')
```

```python
#filtering oldest sports
sports = ['Athletics', 'Cycling', 'Fencing', 'Gymnastics', 'Shooting', 'Swimming', 'Tennis', 'Weightlifting', 'Wrestling']
df = df[(df['Sport'] == sports[0]) | (df['Sport'] == sports[1]) | (df['Sport'] == sports[2]) | (df['Sport'] == sports[3]) | (df['Sport'] == sports[4]) | (df['Sport'] == sports[5]) | (df['Sport'] == sports[6]) | (df['Sport'] == sports[7]) | (df['Sport'] == sports[8])]
```

```python
df_male = df[df['Sex'] == 'M']
df_female = df[df['Sex'] == 'F']
```

```python
# Grouping the data by 'sport', 'medal', and 'country' and counting the occurrences
grouped_data = df_male.groupby(['Sport', 'Medal', 'NOC']).size().reset_index(name='count')

# Sorting the data by count in descending order
sorted_data = grouped_data.sort_values(by='count', ascending=False)

# Selecting only the top 20 rows
top_20_data = sorted_data.head(20)

# Creating a pivot table for better visualization
pivot_table = top_20_data.pivot(index=['Sport', 'Medal'], columns='NOC', values='count')

# Plotting the pivot table
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('Sport - Medal')
plt.ylabel('Count')
plt.title('Top 20 Medal Distribution by Sport and NOC in Male Category')
plt.legend(title='Country')
plt.show()
```

```python
# Grouping the data by 'sport', 'medal', and 'country' and counting the occurrences
grouped_data = df_female.groupby(['Sport', 'Medal', 'NOC']).size().reset_index(name='count')

# Sorting the data by count in descending order
sorted_data = grouped_data.sort_values(by='count', ascending=False)

# Selecting only the top 20 rows
top_20_data = sorted_data.head(20)

# Creating a pivot table for better visualization
pivot_table = top_20_data.pivot(index=['Sport', 'Medal'], columns='NOC', values='count')

# Plotting the pivot table
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('Sport - Medal')
plt.ylabel('Count')
plt.title('Top 20 Medal Distribution by Sport and NOC in Femal Category')
plt.legend(title='Country')
plt.show()
```
