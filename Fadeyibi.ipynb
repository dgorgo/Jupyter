```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('downloads/archive/athlete_events.csv')
df1 = pd.read_csv('downloads/archive/noc_regions.csv')
```

```python
df
```

```python
df2 = pd.merge(df, df1, on="NOC")
```

```python
df2
```

* average number of gold medalists

```python
gold_medalists = df[df['Medal'] == 'Gold']
```

```python
gold_medalists
```

Since we were not given the total number of rows, we calculate it

```python
total_rows = len(df)
```

```python
total_rows
```

```python
average_gold_medalists = len(gold_medalists) / total_rows
```

```python
average_gold_medalists
```

* Gold medalists per city

```python
gold_medalists_by_city = gold_medalists.groupby('City').size()
```

```python
# Plot the average values
gold_medalists_by_city.plot(kind='bar')

# Add labels and title
plt.xlabel('City')
plt.ylabel('Number of Gold Medalists')
plt.title('Gold Medalists By City')
```

* Gold medalists per age

```python
gold_medalists_by_Season = gold_medalists.groupby('Season').size()
```

```python
Seasonal_Gold_Medalists = gold_medalists['Season'].value_counts().sort_values(ascending=False).head(12)

# Plot the average values
gold_medalists_by_Season.plot(kind='pie', autopct='%1.1f%%')

# Add labels and title
plt.title('Gold Medalists By Season')
plt.axis('equal')

plt.show()
Seasonal_Gold_Medalists
```
