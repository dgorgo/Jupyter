```python
# Fill missing values in 'Age' column with the mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values in 'Height' column with the median
df['Height'].fillna(df['Height'].median(), inplace=True)

# Fill missing values in 'Weight' column with the median
df['Weight'].fillna(df['Weight'].median(), inplace=True)
```

### I will only work on summer for the prediction, for top 20 teams. Let's do some data analysis

```python
summer_df.shape
```

### Count of Athletes, Sports and Events

```python
athlete_df = summer_df[['Year','Name','Team','Sport','Event']]

print(athlete_df.shape)
athlete_df.head()
```

```python
athlete_df = athlete_df.groupby(['Year','Team'])[['Name','Sport','Event']]\
                        .nunique()\
                        .reset_index()

print(athlete_df.shape)
athlete_df.head(10)
```

```python
athlete_df = athlete_df.rename(columns={'Name':'Athletes', 'Sport':'Sports', 'Event':'Events'})\
                        .sort_values(['Year','Athletes'],ascending = [True,False])\
                        .reset_index(drop = True)
print(athlete_df.shape)
athlete_df.head(10)
```

### Medal Count

```python
# Selecting only required columns
medals_df = summer_df[['Year','Team','Event','Medal']]

# Drop duplicate rows for team events.
medals_df = medals_df.drop_duplicates(['Team','Year','Event','Medal'])

# Renaming 'Medal column' to 'Medals'
medals_df = medals_df.rename(columns={'Medal': 'Medals'})\
                        .reset_index(drop = True)

print(medals_df.shape)
medals_df.head()
```

```python
medals_df = medals_df[['Year','Team','Medals']]\
                .sort_values(['Year','Team'], ascending = [True,True])\
                .reset_index(drop = True)

print(medals_df.shape)
medals_df.head()
```

```python
medals_df['Gold'] = [1 if cell == 'Gold' else 0 for cell in medals_df['Medals']]
medals_df['Silver'] = [1 if cell == 'Silver' else 0 for cell in medals_df['Medals']]
medals_df['Bronze'] = [1 if cell == 'Bronze' else 0 for cell in medals_df['Medals']]

print(medals_df.shape)
medals_df.head()
```

```python
print(medals_df.shape)
medals_df.head()
```

```python
medals_df = medals_df.drop(['Medals'], axis=1)
print(medals_df.shape)
medals_df.head()
```

```python
# Find total count of medals per year per Team
medals_df = medals_df.groupby(['Year','Team'], sort = False).sum().reset_index()
print(medals_df.shape)
medals_df.head()
```

```python
medals_df['Medals'] = medals_df['Gold'] + medals_df['Silver'] + medals_df['Bronze']

print(medals_df.shape)
medals_df.head()
```

```python
medals_df = medals_df.sort_values(['Year','Medals','Team'], ascending = [True,False,True])

medals_df = medals_df.reset_index(drop=True)

print(medals_df.shape)
medals_df.head()
```

### Merging athele_df and medals_df

```python
dff = athlete_df.merge(medals_df, how ='left').fillna(0)
dff = dff.sort_values(['Year','Medals','Team'], ascending = [True,False,True])

print(dff.shape)
dff.head()
```

```python
dff['Medals'] = dff['Medals'].astype(int)

dff['Gold'] = dff['Gold'].astype(int)

dff['Silver'] = dff['Silver'].astype(int)

dff['Bronze'] = dff['Bronze'].astype(int)

print(dff.shape)
dff.head(10)
```

```python
dff.to_csv('summer_athlete_medals_count.csv',index = False)
```

# Prediction

## Linear Regression to predict Olympic Medal count for top 20 countries (in Summer games)

```python
top_df = pd.read_csv('top_20.csv')
print(top_df.shape)
top_df.head()
```

```python
# Assigning Id to each Team
top_df['TeamId'] = [(x+1) for x in range(20)]

print(top_df.shape)
top_df.head()
```

```python
top_df = top_df[['Team', 'TeamId']]
print(top_df.shape)
top_df.head()
```

```python
df_f = pd.read_csv('summer_athlete_medals_count.csv')
print(df.shape)
df_f.head()
```

```python
# Adding TeamId column to df_f
df_f = pd.merge(df_f,top_df, how = 'left', on = 'Team')
print(df_f.shape)
df_f.head()
```

```python
# Dropping the countries that are not in top 20
df_f = df_f.dropna()
print(df_f.shape)
df_f.head()
```

```python
df_f['Athletes per sport'] = round(df_f['Athletes']/df_f['Sports'], 2)
print(df_f.shape)
df_f.head()
```

```python
# Bring 'TeamId' column to the front
col = df_f.pop('TeamId')
df_f.insert(0,'TeamId', col)

print(df_f.shape)
df_f.head()
```

```python
df_f = df_f.sort_values(['Year', 'Medals', 'Team'], ascending = [True, False, True])
print(df_f.shape)
df_f.head()
```

```python
df_f = pd.get_dummies(df_f, columns = ['Team'])
print(df_f.shape)
df_f.head()
```

### Predicting Medals for 2004

```python
predict_year= 2004
```

#### train data

```python
train_df = df_f[df_f['Year'] < predict_year]
print(train_df.shape)
train_df.head()
```

```python
X = train_df.drop(['Gold', 'Silver', 'Bronze', 'Medals'], axis=1)

y1 = train_df['Gold'].values.reshape(-1, 1)
y2 = train_df['Silver'].values.reshape(-1, 1)
y3 = train_df['Bronze'].values.reshape(-1, 1)
y4 = train_df['Medals'].values.reshape(-1, 1)

print(X.shape, y1.shape, y2.shape, y3.shape, y4.shape)
```

#### Test data

```python
test_df = df_f[df_f['Year'] == predict_year].sort_values(['TeamId'])
print(test_df.shape)
test_df.head()
```

```python
test_data = test_df.drop(['Gold', 'Silver', 'Bronze', 'Medals'], axis = 1).reset_index(drop=True)
print(test_data.shape)
test_data.head()
```

```python
top_df['Gold Actual'] = test_df['Gold'].reset_index(drop=True)
top_df['Silver Actual'] = test_df['Silver'].reset_index(drop=True)
top_df['Bronze Actual'] = test_df['Bronze'].reset_index(drop=True)
top_df['Total Medals Actual'] = test_df['Medals'].reset_index(drop=True)

print(top_df.shape)
top_df.head()
```

#### Test Train Split & Standard Scaler

```python
from sklearn.model_selection import train_test_split

# Gold
X1_train, X1_test, y1_train, y1_test = train_test_split(X, y1, random_state=2)

# Silver
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y2, random_state=2)

# Bronze
X3_train, X3_test, y3_train, y3_test = train_test_split(X, y3, random_state=2)

# Total Medals
X4_train, X4_test, y4_train, y4_test = train_test_split(X, y4, random_state=2)
```

```python
print(X1_train.shape, y1_train.shape, X1_test.shape,  y1_test.shape)
print(X2_train.shape, y2_train.shape, X2_test.shape,  y2_test.shape)
print(X3_train.shape, y3_train.shape, X3_test.shape,  y3_test.shape)
print(X4_train.shape, y4_train.shape, X4_test.shape,  y4_test.shape)
```

#### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

#### Gold

```python
model1 = LinearRegression()

model1.fit(X1_train, y1_train)

training_score1 = model1.score(X1_train, y1_train)
testing_score1 = model1.score(X1_test, y1_test)

print('Gold Medals:')
print(f"Gold Training Score: {training_score1}")
print(f"Gold Testing Score: {testing_score1}")
```

#### Silver

```python
model2 = LinearRegression()

model2.fit(X2_train, y2_train)

training_score2 = model1.score(X2_train, y2_train)
testing_score2 = model1.score(X2_test, y2_test)

print('Silver Medals:')
print(f"Silver Training Score: {training_score2}")
print(f"Silver Testing Score: {testing_score2}")
```

#### Bronze

```python
model3 = LinearRegression()

model3.fit(X3_train, y3_train)

training_score3 = model3.score(X3_train, y3_train)
testing_score3 = model3.score(X3_test, y3_test)

print('Bronze Medals:')
print(f"Bronze Training Score: {training_score3}")
print(f"Bronze Testing Score: {testing_score3}")
```

#### Total Medals

```python
model4 = LinearRegression()

model4.fit(X4_train, y4_train)

training_score4 = model4.score(X4_train, y4_train)
testing_score4 = model4.score(X4_test, y4_test)

print('Total Medals Medals:')
print(f"Total Medals Training Score: {training_score4}")
print(f"Total Medals Testing Score: {testing_score4}")
```

#### Using the models to predict medals for test_data

```python
gold_predictions = model1.predict(test_data)
gold_predictions = np.ravel(gold_predictions)
gold_predictions = np.around(gold_predictions, decimals =0).astype(int)

gold_predictions
```

```python
silver_predictions = model2.predict(test_data)
silver_predictions = np.ravel(silver_predictions)
silver_predictions = np.around(silver_predictions, decimals =0).astype(int)

silver_predictions
```

```python
bronze_predictions = model3.predict(test_data)
bronze_predictions = np.ravel(bronze_predictions)
bronze_predictions = np.around(bronze_predictions, decimals =0).astype(int)

bronze_predictions
```

```python
total_medals_predictions = model4.predict(test_data)
total_medals_predictions = np.ravel(total_medals_predictions)
total_medals_predictions = np.around(total_medals_predictions, decimals =0).astype(int)

total_medals_predictions
```

#### Creating a Dataframe to show all the predictions

```python
top_df['Gold Predicted'] = gold_predictions
top_df['Silver Predicted'] = silver_predictions
top_df['Bronze Predicted'] = bronze_predictions

top_df['Total Medals Predicted'] = total_medals_predictions

top_df['Total Medals Predicted 2'] = top_df['Gold Predicted'] + top_df['Silver Predicted'] + top_df['Silver Predicted']

top_df
```

```python
# Rearranging the columns
top_df = top_df[['Team',\
                 'Gold Actual', 'Gold Predicted',\
                 'Silver Actual','Silver Predicted',\
                 'Bronze Actual', 'Bronze Predicted',\
                 'Total Medals Actual', 'Total Medals Predicted', 'Total Medals Predicted 2' ]]
top_df
```

#### Plot Residuals

```python
y = top_df['Total Medals Actual']

plt.scatter(total_medals_predictions,
            total_medals_predictions - y,
            c = 'red')

plt.hlines(y=0,
           xmin=total_medals_predictions.min(),
           xmax=total_medals_predictions.max(),
           colors ="darkblue")

plt.show()
```

```python
y = top_df['Total Medals Actual']

plt.scatter(top_df['Total Medals Predicted 2'],\
            top_df['Total Medals Predicted 2'] - y,
            c = 'red')

plt.hlines(y=0, xmin=top_df['Total Medals Predicted 2'].min(),\
           xmax=top_df['Total Medals Predicted 2'].max(),\
           colors ="darkblue")

plt.show()
```
